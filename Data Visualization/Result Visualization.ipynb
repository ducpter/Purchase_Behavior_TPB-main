#Result comparision visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a DataFrame
data = {
    'Attitude(ATTD)': [0.84, 0.82, 0.84, 0.84, 0.86, 0.84, 0.86, 0.86],
    'Subjective Naorm (SN)': [0.73, 0.70, 0.84, 0.84, 0.77, 0.77, 0.73, 0.77],
    'Perceived Behavoral Control(PBC)': [0.77, 0.77, 0.80, 0.84, 0.82, 0.77, 0.70, 0.77],
    'PBC & ATTD': [0.86, 0.82, 0.86, 0.86, 0.84, 0.85, 0.84, 0.84],
    'SN & ATTD': [0.89, 0.82, 0.86, 0.86, 0.86, 0.84, 0.86, 0.86],
    'SN & PBC': [0.77, 0.77, 0.80, 0.84, 0.80, 0.77, 0.75, 0.75],
    'ATTD & SN & PBC': [0.84, 0.82, 0.86, 0.86, 0.86, 0.89, 0.86, 0.91]
}

index = ['KNN', 'Decision Tree', 'Logistic Regression', 'Naive Bayes', 'SVM', 'Random Forest', 'Ada Boost', 'Gradient Boosting']

df = pd.DataFrame(data, index=index)
df = df.reset_index().melt('index', var_name='Metric', value_name='Accuracy')

# Initializing the figure and setting the theme
plt.figure(figsize=(10, 8))
sns.set_theme(style="whitegrid")

# Drawing a nested barplot
sns.barplot(x='Accuracy', y='index', hue='Metric', data=df, palette="bright")

# Setting the labels and title
plt.title('Algorithm Accuracies', fontsize=20)
plt.xlabel('Accuracy', fontsize=15)
plt.ylabel('Algorithm', fontsize=15)

# Positioning the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

#Result visualization using Radarplot

```python
#importing libraries
import plotly.graph_objects as go

categories = ['KNN', 'Decision Tree', 'Logistic Regression', 'Naive Bayes', 'SVM', 'Random Forest', 'Ada Boost', 'Gradient Boosting']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[0.73, 0.70, 0.84, 0.84, 0.77, 0.77, 0.73, 0.77],
      theta=categories,
      fill='toself',
      name='Social Naorm (SN)'
))

fig.add_trace(go.Scatterpolar(
      r=[0.77, 0.77, 0.80, 0.84, 0.82, 0.77, 0.70, 0.77],
      theta=categories,
      fill='toself',
      name='Perceived Behavoral Control(PBC)'
))
fig.add_trace(go.Scatterpolar(
      r=[0.86, 0.82, 0.86, 0.86, 0.84, 0.85, 0.84, 0.84],
      theta=categories,
      fill='toself',
      name='Perceived Behavoral Control(PBC) & Attitude(ATTD)'
))
fig.add_trace(go.Scatterpolar(
      r=[0.89, 0.82, 0.86, 0.86, 0.86, 0.84, 0.86, 0.86],
      theta=categories,
      fill='toself',
      name='Social Naorm (SN) & Attitude(ATTD)'
))
fig.add_trace(go.Scatterpolar(
      r=[0.77, 0.77, 0.80, 0.84, 0.80, 0.77, 0.75, 0.75],
      theta=categories,
      fill='toself',
      name='Social Naorm (SN) & Perceived Behavoral Control(PBC)'
))
fig.add_trace(go.Scatterpolar(
      r=[0.84, 0.82, 0.86, 0.86, 0.86, 0.89, 0.86, 0.91],
      theta=categories,
      fill='toself',
      name='Attitude(ATTD) & Social Naorm (SN) & Perceived Behavoral Control(PBC)'
))
fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
)

fig.show()
```
